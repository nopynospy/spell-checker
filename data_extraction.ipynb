```python
recipes_3 = []

import requests
from bs4 import BeautifulSoup
import time
import random

url_head = "https://www.allrecipes.com/recipes/86/world-cuisine/?page="
page_num = 11

for i in range(11, 111):
    resp = requests.get('https://www.allrecipes.com/recipes/86/world-cuisine/?page=' + str(page_num))
    if resp.status_code == 200:
        soup = BeautifulSoup(resp.content, 'html.parser')
        links = soup.find_all("a", {'class' :'tout__imageLink'})
        for link in links:
            new_extract = {
                "name": link['title'],
                "url":link['href']
            }
            recipes_3.append(new_extract)
    page_num += 1
    time.sleep(random.randint(10,15))
```

```python
recipes_3
```

```python
with open('world_recipes_results_3.json', 'w') as f:
    json.dump(recipes_3, f)
```

```python
all_recipes += recipes_3
len(all_recipes)
```
