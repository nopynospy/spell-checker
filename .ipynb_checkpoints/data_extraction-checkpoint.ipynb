After the first ten pages of recipes, the recipes are no longer stored in a dynamic webpage, hence beautifulsoup4 can be used. As the process takes hours and Internet connection issues happen in between, the process has to be restarted after interupted.

```python
import requests

requests.get('https://www.allrecipes.com/recipes/86/world-cuisine/?page=' + str(111))
```

```python
recipes_3 = []

import requests
from bs4 import BeautifulSoup
import time
import random

url_head = "https://www.allrecipes.com/recipes/86/world-cuisine/?page="
page_num = 11

for i in range(11, 111):
    resp = requests.get('https://www.allrecipes.com/recipes/86/world-cuisine/?page=' + str(page_num))
    if resp.status_code == 200:
        soup = BeautifulSoup(resp.content, 'html.parser')
        links = soup.find_all("a", {'class' :'tout__imageLink'})
        for link in links:
            new_extract = {
                "name": link['title'],
                "url":link['href']
            }
            recipes_3.append(new_extract)
    page_num += 1
    time.sleep(random.randint(10,15))
```

```python
recipes_3
```

```python
with open('world_recipes_results_3.json', 'w') as f:
    json.dump(recipes_3, f)
```

```python
all_recipes += recipes_3
len(all_recipes)
```

```python
recipes_4 = []

import requests
from bs4 import BeautifulSoup
import time
import random
import json

url_head = "https://www.allrecipes.com/recipes/86/world-cuisine/?page="
page_num = 11

for i in range(111, 251):
    resp = requests.get('https://www.allrecipes.com/recipes/86/world-cuisine/?page=' + str(i))
    if resp.status_code == 200:
        try:
            soup = BeautifulSoup(resp.content, 'html.parser')
            links = soup.find_all("a", {'class' :'tout__imageLink'})
            for link in links:
                new_extract = {
                    "name": link['title'],
                    "url":link['href']
                }
                recipes_4.append(new_extract)
            with open('recipes_4.json', 'w') as f:
                json.dump(recipes_4, f)
            print(i)
        except Exception as e:
            print(e)
        time.sleep(random.randint(10,15))
    else:
        continue
```
