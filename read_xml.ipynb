```python
import xml.etree.ElementTree as ET
root = ET.parse('FoodBase_curated.xml').getroot()
```

```python
for x in root:
    print(x)
```

```python
full_texts = []

for x in root.findall('document'):
    item =x.find('infon[@key="full_text"]').text
    full_texts.append(item)
    
full_texts
```

```python
len(full_texts)
```

```python
sum( map(len, full_texts) ) / len(full_texts)
```

```python
corpus = ' '.join(full_texts)
corpus = corpus.replace("\n\t\t", '')
corpus
```

```python
import nltk

tokens = nltk.word_tokenize(corpus)
len(tokens)
```

```python
bigrams = nltk.bigrams(tokens)
frequence = nltk.FreqDist(bigrams)
for key,value in frequence.items():
    print(key,value)
```

```python
unique_tokens = sorted(list(set(tokens)))
unique_tokens
```

```python
len(unique_tokens)
```

```python
result = dict(sorted(frequence.items(), key=lambda item: item[0]))
result
```

```python
def find_dict_tuple_key(search):
    entry = {
        "token": search,
        "bigrams": []
    }
    bigrams = {x: result[x] for x in result.keys() if x[0] == search}
    for key, value in bigrams.items():
        newDict = {key[1]: value}
        entry["bigrams"].append(newDict)
    return entry

find_dict_tuple_key('!')
```

```python
bigram_list = []
for value in unique_tokens:
    bigram_list.append(find_dict_tuple_key(value))
bigram_list
```

```python
len(bigram_list)
```

# Pre-processing

```python
import re

def searchWordsPatt(str, patt):
    array = re.findall(patt, str)
    return array

NUMPATTERN = r'[0-9]+'
sorted(searchWordsPatt(corpus, NUMPATTERN))
```

```python
newCorpus = re.sub(NUMPATTERN, '4', corpus)
searchWordsPatt(newCorpus, NUMPATTERN)
```

```python
tokens = nltk.word_tokenize(newCorpus)
bigrams = nltk.bigrams(tokens)
frequence = nltk.FreqDist(bigrams)
for key,value in frequence.items():
    print(key,value)
```

```python
unique_tokens = sorted(list(set(tokens)))
unique_tokens
```

```python
len(unique_tokens)
```

```python
result = dict(sorted(frequence.items(), key=lambda item: item[0]))
result
```

```python
bigram_list = []
for value in unique_tokens:
    bigram_list.append(find_dict_tuple_key(value))

bigram_list
```
